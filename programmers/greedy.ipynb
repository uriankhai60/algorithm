````python
def solution(n, lost, reserve):
    ss = [1]*n
    
    for l in lost:
        ss[l-1]-=1
    for r in reserve:
        ss[r-1]+=1


    for idx, s in enumerate(ss):
        if s==0:
            ## 여기 조건에서 틀림 idx=1일때도 왼쪽에게 못빌리는 현상이 발생
            if (idx-1 > 0) and (ss[idx-1] == 2): 
                ss[idx-1] -= 1
                ss[idx] += 1
                continue
            
            elif (idx+1 < n) and (ss[idx+1] == 2):
                ss[idx+1] -= 1
                ss[idx] += 1
                continue
        else:
            pass
        
    answer = sum([True for i in ss if i > 0])
    return answer
````

````python
'''
오답

스택을 2개를 만듬.
버릴스택, 답안으로 제출할 스택(=답안스택)

입력 스택에서 하나씩 팝을 하고
답안으로 제출할 스택의 마지막과 비교함

입력스택 팝의 값이 더 크면,
답안스택 꺼내서 버릴스택에 넣고
팝의 값을 답안스택에 넣는다

입력스택 팝의 값이 같으면
답안스택에 넣는다

입력스택 팝의 값이 작으면
버릴스택에 넣는다.
'''

from collections import deque

def solution(number, k):
    del_stack = []
    new_stack = []
    dq = deque(number)
    new_stack.append(dq.popleft())
    while dq:
        cur = dq.popleft()
        if new_stack[-1] > cur:
            del_stack.append(cur)
        elif new_stack[-1] == cur:
            new_stack.append(cur)
        else: # new_stack[-1] < cur
            pre = new_stack.pop()
            del_stack.append(pre)
            new_stack.append(cur)
            
        if len(del_stack) == k:
            nxt = dq.popleft()
            last = del_stack.pop()
            # 만약 다음게 큰 경우
            if last > nxt:
                del_stack.append(nxt)
                dq.appendleft(last)
            else:
                del_stack.append(last)
                dq.appendleft(nxt)
            break
    new_stack = new_stack + list(dq)
    answer = ''.join(new_stack)
    return answer

'''
K개를 반드시 지운다라는 로직이 없음
예를들어 (9990, 2)이 들어온경우
답은 999가 됨

del_stack == k 이전에 dq가 비어비리면
논리상으로 문제가 될 수 있음

또한 마지막 보정로직도 순서를 변경해버리기 때문에
붙여서 해결되지 않음
'''
````

````python
'''
앞에있는 숫자가 나보다 작다면 지우는게 이득이다.
'''

def solution(number, k):
    answer = ''
    
    stack = [] # 답안
    remove_cnt = k
    
    # 앞부분부터 루프
    for digit in number:
        
        # 스택에 무엇인가가 있고
        # 리무브 카운트가 남아있고
        # 스택 마지막보다 digit이 크면
        while len(stack)>0 and remove_cnt>0 and stack[-1] < digit:
            stack.pop() # 저장되어 있는 값 스택에서 제거
            remove_cnt -= 1 # 카운트 증가
        
        # 스택에 값은 무조건 저장
        stack.append(digit)
     
    # 리무브 카운트가 남아있다면 뒤에서부터 잘라낸다
    if remove_cnt > 0:
        stack = stack[:-remove_cnt]
            
    
    return ''.join(stack)
````
