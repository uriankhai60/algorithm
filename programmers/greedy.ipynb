## 체육복(0,1)  
## 기억할 것   
* 초기 조건은 항상 꼼꼼하게 볼것  
## 링크  
[문제 링크](https://school.programmers.co.kr/learn/courses/30/lessons/42862)  

```python
def solution(n, lost, reserve):
    ss = [1]*n
    
    for l in lost:
        ss[l-1]-=1
    for r in reserve:
        ss[r-1]+=1


    for idx, s in enumerate(ss):
        if s==0:
            ## 여기 조건에서 틀림 idx=1일때도 왼쪽에게 못빌리는 현상이 발생
            if (idx-1 > 0) and (ss[idx-1] == 2): 
                ss[idx-1] -= 1
                ss[idx] += 1
                continue
            
            elif (idx+1 < n) and (ss[idx+1] == 2):
                ss[idx+1] -= 1
                ss[idx] += 1
                continue
        else:
            pass
        
    answer = sum([True for i in ss if i > 0])
    return answer
```
